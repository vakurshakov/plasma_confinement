```python
phase = []
with open("phase_data.txt", 'r') as f:
    pmin_pmax = f.readline().split(' ')
    pmin = (int)(pmin_pmax[0])
    pmax = (int)(pmin_pmax[1])
    хmin_хmax = f.readline().split(' ')
    xmin = (int)(хmin_хmax[0])
    xmax = (int)(хmin_хmax[1])
    
    for t in range(1):
        temp = f.readline() 
        temp = temp.split('\t')[:-1]
        phase_X = []
        for X in range(len(temp)) : # [0.01, 0.02, 0.03]
            phase_X.append(list(map(float, temp[X].split(' ')[:-1])))
        phase.append(phase_X)
```

```python
######### / animation /########################################################
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from mpl_toolkits.axes_grid1 import make_axes_locatable
%matplotlib notebook


fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot(111, xlim=(xmin, xmax),  ylim=(pmin, pmax))


phase_ = ax.imshow(phase[0],
                cmap = plt.get_cmap('plasma'),
                origin='lower',
                extent=(xmin,xmax,pmin,pmax),
                #vmin=-0.127, vmax=0.127,
                )

fig.colorbar(phase_,
             cax=make_axes_locatable(ax).append_axes("right", 
                                                     size="4%", 
                                                     pad=0.05))
plt.show()
```
